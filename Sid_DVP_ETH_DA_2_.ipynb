```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
url = "https://raw.githubusercontent.com/ishanjogalekar/DataSets/main/police_dataset.csv"
df = pd.read_csv(url)
df.head()
```

**Box plot**

```
sns.set_theme(style="whitegrid")
ax = sns.boxplot(x=df['age'])
```

**Barplot**

```
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10,10))
sns.barplot(data=df,y='age',x='armed').set(title = "Armed vs Age")
```

**Violine Plot**

```
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10,10))
sns.violinplot(data=df,y='age',x='armed').set(title = "Armed vs Age")
```

**KDE-plot**

```
plt.figure(figsize=(10,10))
df.plot(kind='density', subplots=True, layout=(3,3), sharex=False , figsize =(15,15))
plt.show()
```

**Pie-chart**

```
F_gender = df.gender[df.gender == 'F'].count()
M_gender = df.gender[df.gender == 'M'].count()
gender_nums = [F_gender, M_gender]
labels = ["Female", "Male"]
pie_colors = ['lightblue', 'red']
plt.figure(figsize=(5,5))
plt.pie(gender_nums, labels = labels, colors=pie_colors,autopct='%1.1f%%')
plt.title("Pie Chart by gender", fontweight='bold', fontsize=22, color='grey')
plt.axis('equal')
plt.show()
```
