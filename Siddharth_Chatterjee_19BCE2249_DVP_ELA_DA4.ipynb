### Importing libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Ignore Warnings
import warnings
warnings.filterwarnings("ignore")
```

### Seaborn settings

```python
sns.set(style="white",color_codes=True)
```

### Dataset - Tips

```python
url = "https://raw.githubusercontent.com/siddharthchatterjee9/Data-Vizualization-and-Presentation/main/tips.csv"
df =pd.read_csv(url) 
df.head()
```

```python
df.describe()
```

```python
df.columns
```

### Plots
1. Scatter Plot
2. Joint plot
3. Box plot
4. Strip plot
5. Violin plot
6. KDE Plot
7. Pair plot

### Scatter plot

```python
df.plot(kind="scatter",x="day",y= "total_bill")
plt.show()
```

```python
sns.FacetGrid(df,hue='day',size=5) \
    .map(plt.scatter,"day","total_bill") \
    .add_legend()
```

### Joint plot

```python
sns.jointplot(x="day",y="total_bill",data=df,size=5)
plt.show()
```

### Box plot

```python
sns.boxplot(x='day',y='tip',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df)
```

```python
df.boxplot(by="day")
```

### Strip plot

```python
sns.stripplot(x="day",y="tip",data=df,size=6)
```

```python
sns.stripplot(x="day",y="total_bill",data=df,size=6)
```

```python
sns.stripplot(x="day",y="size",data=df,size=6)
```

### Violin plot

```python
sns.violinplot(x="day",y="tip",data=df,size=6)
```

```python
sns.violinplot(x="day",y="total_bill",data=df,size=6)
```

```python
sns.violinplot(x="sex",y="tip",data=df,size=6)
```

```python
sns.violinplot(x="sex",y="total_bill",data=df,size=6)
```

### KDE plot

```python
sns.FacetGrid(df,hue='day',height=5) \
    .map(sns.kdeplot,"total_bill") \
    .add_legend()
```

```python
sns.FacetGrid(df,hue='sex',height=5) \
    .map(sns.kdeplot,"tip") \
    .add_legend()
```

```python
sns.FacetGrid(df,hue='sex',height=5) \
    .map(sns.kdeplot,"total_bill") \
    .add_legend()
```

### Pair plot

```python
sns.pairplot(data=df, hue="day", height=3)
```

```python
sns.pairplot(data=df, hue="sex", height=3)
```
