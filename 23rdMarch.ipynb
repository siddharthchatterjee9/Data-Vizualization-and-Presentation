```python
import pandas as pd
from plotnine import *
```

```python
url = "https://raw.githubusercontent.com/ishanjogalekar/DataSets/main/insurance.csv";
data = pd.read_csv(url)
```

```python
data.head()
```

```python
data.describe()
```

```python
regions = ['southwest', 'northwest', 'northeast', 'southeast']
```

```python
from plotnine.scales.scale_xy import scale_x_discrete
(
    ggplot(data)
    + geom_boxplot(aes(x='factor(region)', y='bmi', color='region'))
    + scale_x_discrete(labels=regions, name=regions)
    + theme_538()
)
```

```python
(
    ggplot(data, aes(x = 'charges', y = 'bmi', color = 'smoker'))
    + geom_point()
    + facet_grid('children ~ .')
    + labs(x = 'charges', y = 'bmi')
    + theme_538()
)
```

```python
(
    ggplot(data, aes(x = 'charges', color = 'smoker'))
    + geom_histogram()
)
```

```python
(
    ggplot(data, aes(x='charges', color='smoker'))
    + geom_histogram(fill='white')
    + facet_grid('sex ~ region')
)
```

```python
(
    ggplot(data)
    + geom_count(mapping = aes(x = 'region', y = 'sex', color = 'region'))
)
```
