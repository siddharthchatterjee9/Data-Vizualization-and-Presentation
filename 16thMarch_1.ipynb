```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
url = "https://raw.githubusercontent.com/ishanjogalekar/DataSets/main/dataset_lab.csv"
df = pd.read_csv(url)
df.drop(df.columns[df.columns.str.contains('Unnamed',case = False)],axis = 1, inplace = True)
```

```python
df.drop(df.columns[df.columns.str.contains('Feature Name', case=False)], axis=1, inplace=True)
df.head()
```

```python
fig = px.bar(df, x='Size')
fig.show()
```

```python
fig = px.bar(df,  x='Feature Name',  y='Size', color='Feature Name')
fig.show()
```

```python
url_new = "https://raw.githubusercontent.com/ishanjogalekar/DataSets/main/pune_1965_to_2002.csv"
data = pd.read_csv(url_new)
data.head()
```

```python
data.set_index("Year", inplace=True)
```

```python
norm = np.log2(data)
```

```python
data.head()
```

```python
fig = px.imshow(norm, color_continuous_scale='RdBu_r')
fig.show()
```

```python
fig = px.box(norm)
fig.show()
```

```python
fig = px.line(data, title = 'Precipitation vs Year (which connects months)')
fig.show()
```

```python
data1 = data.transpose()
fig = px.line(data1, title='Precipitation vs Months')
fig.show()
```
