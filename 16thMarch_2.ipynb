```python
import pandas as pd
import plotly.express as px
```

```python
url = "https://raw.githubusercontent.com/ishanjogalekar/DataSets/main/CasesPerCountry.csv"
df = pd.read_csv(url)
df.head()
```

```python
df2 = df.drop(['FIPS', 'Admin2', 'Province_State', 'Recovered', 'Active'], axis=1)
df2.head()
```

```python
fig = px.scatter(df2, x="Country_Region", y="Case_Fatality_Ratio")
fig.show()
```

```python
data3 = px.data.gapminder()
data3.head()
```

```python
fig4 = px.scatter(data3.query("year==2007"), x="gdpPercap", y="lifeExp", size="pop", color="continent", hover_name="country", log_x=True, size_max=60)
fig4.show()
```
